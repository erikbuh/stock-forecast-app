```python
import yfinance as yf
from prophet import Prophet
import pandas as pd
```

```python
# load data --> microsoft stock price
data = yf.download('msft', period="1y", interval="1d")
```

```python
# pandas dataframe table
data.tail(10)
```

```python
# new dataframe with only Date and Close
df = data.reset_index()[['Date', 'Close']]
# rename thes eto ds and y
df.columns = ['ds', 'y']
# remove time zone information  --> only dates
df['ds'] = df['ds'].dt.tz_localize(None)
```

```python
df.tail(10)
```

```python
# Train Prophet model
model = Prophet(daily_seasonality=True)
model.fit(df)
```

```python
# Make future predictions
periods = 5 # Number of days to forecast

future = model.make_future_dataframe(periods=periods)
forecast = model.predict(future)
```

```python
forecast.tail(20)
```

```python
# remove a lot of columns
forecast = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```
